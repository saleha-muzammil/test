```python
# Creating a list
my_list = [1, 2, 3, 4, 5]

# Accessing an element
my_list[0] # returns the first element in the list, which is 1

# Adding elements to a list
my_list.append(6) # adds 6 to the end of the list
my_list.insert(0, 0) # adds 0 to the beginning of the list

# Removing elements from a list
my_list.remove(3) # removes the first occurrence of 3 from the list
my_list.pop(0) # removes the first element from the list

# Slicing a list
my_list[1:3] # returns a new list with elements at index 1 and 2
my_list[:3] # returns a new list with elements at index 0, 1, and 2
my_list[3:] # returns a new list with elements starting from index 3

# Sorting a list
my_list.sort() # sorts the list in ascending order
my_list.reverse() # reverses the order of the list

# Checking if an element is in a list
3 in my_list # returns True if 3 is in the list

# Finding the length of a list
len(my_list) # returns the number of elements in the list

# Copying a list
new_list = my_list.copy() # creates a new list with the same elements as my_list

# Clearing a list
my_list.clear() # removes all elements from the list
```

```python
# Creating a dictionary
my_dict = {'key1': 'value1', 'key2': 'value2'}

# Accessing a value using a key
my_dict['key1'] # returns 'value1'

# Adding a key-value pair to a dictionary
my_dict['key3'] = 'value3'

# Modifying a value in a dictionary
my_dict['key1'] = 'new_value1'

# Removing a key-value pair from a dictionary
del my_dict['key2']
my_dict.pop('key3')

# Checking if a key is in a dictionary
'key1' in my_dict # returns True if 'key1' is in the dictionary

# Getting the number of key-value pairs in a dictionary
len(my_dict)

# Getting a list of keys or values in a dictionary
my_dict.keys()
my_dict.values()

# Looping through a dictionary
for key in my_dict:
    print(key, my_dict[key])

# Copying a dictionary
new_dict = my_dict.copy()

# Clearing a dictionary
my_dict.clear()
```

```python
# Creating a set
my_set = {1, 2, 3}

# Adding elements to a set
my_set.add(4)
my_set.update([5, 6, 7])

# Removing elements from a set
my_set.remove(3)
my_set.discard(4)
my_set.pop() # removes and returns an arbitrary element from the set

# Set operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

set3 = set1.union(set2) # returns a new set with all elements from set1 and set2
set4 = set1.intersection(set2) # returns a new set with elements common to set1 and set2
set5 = set1.difference(set2) # returns a new set with elements in set1 but not in set2
set6 = set1.symmetric_difference(set2) # returns a new set with elements in set1 or set2 but not both

# Checking if a set is a subset or superset of another set
set1.issubset(set3) # returns True if set1 is a subset of set3
set3.issuperset(set1) # returns True if set3 is a superset of set1

# Checking if two sets are disjoint
set1.isdisjoint(set2) # returns False if set1 and set2 have at least one common element

# Getting the length of a set
len(my_set)

# Copying a set
new_set = my_set.copy()

# Clearing a set
my_set.clear()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
import re

property_df = pd.read_csv('HousePredictionProject (1).csv')
property_df.info()


property_df = property_df.reset_index()
property_df = property_df.drop("index",axis=1)
property_df.head()

print(property_df["purpose"].unique())
print(property_df["property_type"].unique())
print(property_df["location"].unique())
print(property_df["city"].unique())


property_df = property_df[property_df["city"] == "Lahore"]
property_df = property_df[property_df["property_type"] == "House"]
print(property_df.head())

df = property_df[['price', 'location', 'baths','area','bedrooms']]

df.info()
print(df)

# Checking if any null values in the df
df.isna().sum()

df = df.drop(df[(df['baths']==0) & (df['bedrooms'] > 3)].index)
df = df.drop(df[(df['baths']==0) & (df['bedrooms'] == 0)].index)
df.drop(df[(df['bedrooms']==0) | (df['baths']==0)].index, inplace=True)
df[df["baths"] > df["bedrooms"]]
df.drop(df[df['baths'] > df['bedrooms']].index, inplace=True)
print(df)
print(df)

def sqft(area):
    area = str(area)
    if 'Marla' in area:
        area = float(re.sub(r'\D+', '', area)) * 272.251
    elif 'Kanal' in area:
        area = float(re.sub(r'\D+', '', area)) * 5445
    return area

df['area'] = df['area'].apply(sqft)
print(df)

df['area'] = pd.to_numeric(df['area'])
df['price_per_sqft'] = df['price'] / df['area']

final = df[['price', 'location', 'baths','bedrooms','price_per_sqft']]

# Save the output to a new file
final.to_csv('output.csv', index=False)

d= pd.read_csv('output.csv')
d.info()
```
